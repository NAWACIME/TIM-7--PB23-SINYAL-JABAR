```python
# Memilih fitur untuk clustering sesuai data sinyal.csv
features = ['SINYAL KUAT',
            'SINYAL LEMAH',
            'TIDAK ADA SINYAL']
```

```python
# Standardisasi data sesuai kolom sinyal.csv
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[['SINYAL KUAT',
                                           'SINYAL LEMAH',
                                           'TIDAK ADA SINYAL']])
```

```python
# Menentukan jumlah cluster menggunakan metode Elbow
inertia = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(scaled_features)   # <-- ini sudah menggunakan Sinyal.csv
    inertia.append(kmeans.inertia_)
```

```python
# Plot Elbow Method menggunakan data sinyal.csv
plt.figure(figsize=(8, 5))
plt.plot(k_values, inertia, marker='o')
plt.title("Elbow Method untuk Menentukan Jumlah Cluster Optimal (Data Sinyal Desa)")
plt.xlabel("Jumlah Cluster (k)")
plt.ylabel("Inertia")
plt.grid()
plt.show()
```

```python
# Menggunakan jumlah cluster optimal
optimal_k = 4 # Berdasarkan elbow method (misalnya)
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init='auto')
df['cluster'] = kmeans.fit_predict(scaled_features)
```

```python
# Mengurangi dimensi data untuk visualisasi
pca = PCA(n_components=2)
reduced_features = pca.fit_transform(scaled_features)
```

```python
# 1. Pastikan cluster disimpan
df['cluster'] = kmeans.fit_predict(scaled_features)

# 2. Setup Plot
plt.figure(figsize=(10, 7))

# Menggunakan colormap otomatis agar tidak perlu list 'colors' manual
# 'tab10' sangat bagus untuk kategorisasi cluster
cmap = plt.get_cmap('tab10')

for cluster in range(optimal_k):
    # Filter data berdasarkan cluster saat ini
    cluster_points = reduced_features[df['cluster'] == cluster]
    
    plt.scatter(cluster_points[:, 0],
                cluster_points[:, 1],
                s=50,
                label=f'Cluster {cluster}',
                color=cmap(cluster % 10), # Mengambil warna otomatis dari cmap
                alpha=0.7) # Sedikit transparan agar titik yang tumpang tindih terlihat

# 3. Plot Centroids
centroids_reduced = pca.transform(kmeans.cluster_centers_)
plt.scatter(centroids_reduced[:, 0],
            centroids_reduced[:, 1],
            s=250,        # Ukuran lebih besar agar mencolok
            c='black',    # Warna hitam kontras
            marker='X',   # Bentuk X
            edgecolor='white', # Garis tepi putih agar jelas
            label='Centroids')

# 4. Dekorasi Plot
plt.title("Visualisasi Clustering K-Means (PCA 2D)", fontsize=14, pad=15)
plt.xlabel("Principal Component 1 (PC1)")
plt.ylabel("Principal Component 2 (PC2)")
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') # Legend di luar plot agar rapi
plt.tight_layout()
plt.show()
```
